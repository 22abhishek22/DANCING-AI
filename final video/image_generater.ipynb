```python
# Creating Pairs to connect coordinates
POSE_PAIRS = [[0,1], [1,2], [2,3], [3,4], [1,5], [5,6], [6,7], [2,8], [8,9], [9,10], [5,11], [11,12], [12,13], [8,11]]
```

```python
# Creating VideoWriter object
vid_writer = cv2.VideoWriter('output.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame.shape[1],frame.shape[0]))
```

```python
# Draw Skeleton

for ij in range(0,300):
    # Reading background image for video
    frame = cv2.imread('black.jpg')
    
    # Creating a tuple and a list to read coordinates from dataframe
    coor = ()
    points = []
    
    # Storing coordinates in the list points
    for i in range(0,36,2):
        coor = (int(df.iloc[ij,i]),) + (int(df.iloc[ij,i+1]),)
        points.append(coor)

    # Generating AI boody
    for pair in POSE_PAIRS:
        partA = pair[0]
        partB = pair[1]

        if points[partA] and points[partB]:
            cv2.line(frame, points[partA], points[partB], (191, 201, 202), 3, lineType = 8)
            cv2.circle(frame, points[partA], 7, (0, 0, 150), thickness=-1, lineType=cv2.FILLED)
            cv2.circle(frame, points[partB], 7, (0, 0, 150), thickness=-1, lineType=cv2.FILLED)
    
    # Showing generated frames
    cv2.imshow('Output-Skeleton', frame)
    
    # Writing frames to vid_writer object
    vid_writer.write(frame)

# Releasing Final video
vid_writer.release()
```
