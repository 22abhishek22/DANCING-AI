```python
# Importing required libraries

import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For displaying all columns of a dataframe

pd.set_option('display.max_columns',None)
```

```python
# Converting audio to a floating point series
# Audio is same as that in video
# y is the audio point series

y, sr = librosa.load('sample_video.wav', duration = 58)

# sr is the sample rate
sr
```

```python
#Printing shape of the floating point series

y.shape
```

```python
# Converting shape of y to the shape of video frames
# hop_length is the number of audio samples between successive onset measurements
# length of audio and video = 58 sec, Frames in video = 1750

y1 = librosa.feature.tempogram(y, sr, hop_length = int((58*sr)/1750), win_length = 36)
```

```python
print(y1.shape)

# Converting y1 to dataframe
y1 = pd.DataFrame(y1)
y1.head()
```

```python
# Storing dataframe y1 to audio.csv file
y1.to_csv('audio.csv',index = None)
```
